```python
#create path where file located
path1 = "/Users/nirosh/Downloads/pp/data/aggregated/transaction/country/india/state"
agg_trans_list = os.listdir(path1)
columns1 = {'State': [], 'Year': [], 'Quarter': [], 'Transaction_type': [], 'Transaction_count': [],
            'Transaction_amount': []}
for state in agg_trans_list:
    cur_state = path1 +'/'+ state + "/"
    agg_year_trans = os.listdir(cur_state)
    
    for year in agg_year_trans:
        cur_year = cur_state + year + "/"
        agg_file_trans = os.listdir(cur_year)
        
        for file in agg_file_trans:
            cur_file = cur_year + file
            data = open(cur_file, 'r')
            A = json.load(data)
            
            for i in A['data']['transactionData']:
                name = i['name']
                count = i['paymentInstruments'][0]['count']
                amount = i['paymentInstruments'][0]['amount']
                columns1['Transaction_type'].append(name)
                columns1['Transaction_count'].append(count)
                columns1['Transaction_amount'].append(amount)
                columns1['State'].append(state)
                columns1['Year'].append(year)
                columns1['Quarter'].append(int(file.strip('.json')))
                
df_agg_trans = pd.DataFrame(columns1)
```

```python
#df_agg_trans.isnull().sum() =====> there is no null values and data is cleaned
#df_agg_trans.duplicated() ====> No duplicates found
df_agg_trans['Year'] = df_agg_trans['Year'].astype(int)

# Details of agg users
path2 = '/Users/nirosh/Downloads/pp/data/aggregated/user/country/india/state'
agg_user_list = os.listdir(path2)
columns_two = {'State':[],'Year':[],'Quarter':[],'Brand':[],'Count':[],'Percentage':[]}

for state in agg_user_list:
    cur_state = path2+'/'+state+'/'
    agg_year_user = os.listdir(cur_state)
    
    for year in agg_year_user:
        cur_year = cur_state+year+'/'
        agg_file_user = os.listdir(cur_year)
        
        for file in agg_file_user:
            cur_file = cur_year+file
            data = open(cur_file,'r')
            B = json.load(data)
            
            try:
                for i in B['data']['usersByDevice']:
                    brand = i['brand']
                    count = i['count']
                    percent = i['percentage']
                    columns_two['Brand'].append(brand)
                    columns_two['Count'].append(count)
                    columns_two['Percentage'].append(percent)
                    columns_two['State'].append(state)
                    columns_two['Year'].append(year)
                    columns_two['Quarter'].append(int(file.strip('.json')))
            except:
                pass
df_agg_user = pd.DataFrame(columns_two)
#df_agg_user.isnull().sum()===> No null values , so Data is cleaned
#df_agg_user.duplicated().unique() ===> no duplicates
df_agg_user['Year'] = df_agg_user['Year'].astype(int)

# Details of map transactions
path3 = '/Users/nirosh/Downloads/pp/data/map/transaction/hover/country/india/state'
map_trans_list = os.listdir(path3)
columns_three = {'State':[],'Year':[],'Quarter':[],'District':[],'Count':[],'Amount':[]}

for state in map_trans_list:
    cur_state = path3+'/'+state+'/'
    map_year_trans = os.listdir(cur_state)
    
    for year in map_year_trans:
        cur_year = cur_state+year+'/'
        map_file_trans = os.listdir(cur_year)
        
        for file in map_file_trans:
            cur_file = cur_year+file
            data = open(cur_file,'r')
            C = json.load(data)
            
            for i in C['data']['hoverDataList']:
                district = i['name']
                count = i['metric'][0]['count']
                amount = i['metric'][0]['amount']
                columns_three['District'].append(district)
                columns_three['Count'].append(count)
                columns_three['Amount'].append(amount)
                columns_three['State'].append(state)
                columns_three['Year'].append(year)
                columns_three['Quarter'].append(int(file.strip('.json')))
                
df_map_trans = pd.DataFrame(columns_three)
                
                
df_map_trans #data is clean and no duplicates found
df_map_trans['Year'] = df_map_trans['Year'].astype(int)

# Details of map user
path4 = '/Users/nirosh/Downloads/pp/data/map/user/hover/country/india/state'
map_user_list = os.listdir(path4)
columns_four = {'State':[],'Year':[],'Quarter':[],'District':[],'RegisteredUsers':[],'Appopens':[]}

for state in map_user_list:
    cur_state = path4+'/'+state+'/'
    map_year_user = os.listdir(cur_state)
    
    for year in map_year_user:
        cur_year = cur_state+year+'/'       
        map_file_user = os.listdir(cur_year)
        
        for file in map_file_user:
            cur_file = cur_year+file
            data = open(cur_file,'r')
            D = json.load(data)
            
            for i in D['data']['hoverData'].items():
                district = i[0]
                users = i[1]['registeredUsers']
                appopens = i[1]['appOpens']
                columns_four['State'].append(state)
                columns_four['Year'].append(year)
                columns_four['Quarter'].append(int(file.strip('.json')))
                columns_four['District'].append(district)
                columns_four['RegisteredUsers'].append(users)
                columns_four['Appopens'].append(appopens)
                
df_map_user = pd.DataFrame(columns_four)

                
df_map_user #data is clean, so no duplicates found
df_map_user['Year'] = df_map_user['Year'].astype(int)

# Details Of Top Transactions
path5 = '/Users/nirosh/Downloads/pp/data/top/transaction/country/india/state'
top_trans_list = os.listdir(path5)
columns_five = {'State': [], 'Year': [], 'Quarter': [], 'Pincode': [], 'Transaction_count': [],
            'Transaction_amount': []}

for state in top_trans_list:
    cur_state = path5 + '/'+state + "/"
    top_year_trans = os.listdir(cur_state)
    
    for year in top_year_trans:
        cur_year = cur_state + year + "/"
        top_file_trans = os.listdir(cur_year)
        
        for file in top_file_trans:
            cur_file = cur_year + file
            data = open(cur_file, 'r')
            E = json.load(data)
            
            for i in E['data']['pincodes']:
                name = i['entityName']
                count = i['metric']['count']
                amount = i['metric']['amount']
                columns_five['Pincode'].append(name)
                columns_five['Transaction_count'].append(count)
                columns_five['Transaction_amount'].append(amount)
                columns_five['State'].append(state)
                columns_five['Year'].append(year)
                columns_five['Quarter'].append(int(file.strip('.json')))
df_top_trans = pd.DataFrame(columns_five)
df_top_trans['Pincode']=df_top_trans['Pincode'].fillna(194109)
df_top_trans['Pincode'] = df_top_trans['Pincode'].astype(int)
df_top_trans['Year'] = df_top_trans['Year'].astype(int)
```
